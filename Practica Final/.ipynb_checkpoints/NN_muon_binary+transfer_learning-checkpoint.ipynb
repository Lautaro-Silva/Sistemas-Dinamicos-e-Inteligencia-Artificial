```python
import torch
from torch.utils.data import TensorDataset, DataLoader, random_split
import torch.nn.functional as F
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# -----------------------
# CONFIG
# -----------------------
pulse_samples = 1000     # n_feat_muon
n_feat_frec = 2500       # n_feat_frec
num_muones_dummy = 16    # para probar forward

# Paths a tus datasets
muon_dataset_path = "muon_dataset_segmented_v3.npz"
simulated_npz_path = "Espectros/spectrum_simulated.npz"
real_npz_path = "Espectros/spectrum_real.npz"

# -----------------------
# CARGA DE DATASET DE MUONES
# -----------------------
data = np.load(muon_dataset_path)
X_muones = data['X']        # shape (num_pulsos, pulse_samples)
y_muones = data['y']        # shape (num_pulsos,)

# Convertir a tensores PyTorch
X_muones = torch.tensor(X_muones, dtype=torch.float32)
y_muones = torch.tensor(y_muones, dtype=torch.float32).unsqueeze(-1)  # (N,1)

print(f"Señales de muones: {X_muones.shape}, targets: {y_muones.shape}")

# -----------------------
# CARGA DE ESPECTROS
# -----------------------
def load_npz(path):
    data = np.load(path)
    freq = data['freq']
    att = data['attenuation_db']
    return torch.tensor(freq, dtype=torch.float32), torch.tensor(att, dtype=torch.float32)

sim_freq, H_sim = load_npz(simulated_npz_path)
real_freq, H_real = load_npz(real_npz_path)

# -----------------------
# DEFINICIÓN DE LA RED
# -----------------------
class MuonFrequencyNet(nn.Module):
    def __init__(self, n_feat_muon=1000, n_feat_frec=2500):
        super(MuonFrequencyNet, self).__init__()

        # -----------------------
        # Rama 1: Pulsos de muones (CNN 1D)
        # -----------------------
        self.cnn_branch = nn.Sequential(
            nn.Conv1d(1, 32, kernel_size=10, padding='same'),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2),

            nn.Conv1d(32, 64, kernel_size=8, padding='same'),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2),

            nn.Conv1d(64, 128, kernel_size=5, padding='same'),
            nn.ReLU(),
            nn.AdaptiveAvgPool1d(1)  # Global Average Pooling
        )

        # -----------------------
        # Rama 2: Respuesta en frecuencia simulada (MLP)
        # -----------------------
        self.mlp_branch = nn.Sequential(
            nn.Linear(n_feat_frec, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU()
        )

        # -----------------------
        # Cabezal de fusión
        # -----------------------
        self.head = nn.Sequential(
            nn.Linear(128 + 64, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 1)  # Output: número de #1s normalizado
        )

    def forward(self, x_muon, x_freq):
        """
        x_muon: (batch_size, 1, n_feat_muon)
        x_freq: (batch_size, n_feat_frec)
        """
        # Rama 1
        v1 = self.cnn_branch(x_muon)
        v1 = v1.view(v1.size(0), -1)  # Flatten

        # Rama 2
        v2 = self.mlp_branch(x_freq)

        # Concatenar
        merged = torch.cat([v1, v2], dim=1)

        # Cabezal
        out = self.head(merged)
        return out
        
# -----------------------
# Inicializar modelo
# -----------------------
model = MuonFrequencyNet(n_feat_muon=pulse_samples, n_feat_frec=n_feat_frec)
print(model)
```

```python
# -----------------------
# Probar forward con batch dummy
# -----------------------
batch_muon = X_muones[:num_muones_dummy]
batch_freq = H_sim.repeat(num_muones_dummy, 1)  # repetir espectro simulado para cada ejemplo
output = model(batch_muon, batch_freq)
print("Output shape:", output.shape)  # debe ser (num_muones_dummy, 1)

# Listo para entrenamiento con optimizer y loss
```

```python
# -----------------------
# CONFIG
# -----------------------
batch_size = 32
epochs = 30
lr = 1e-3
device = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------
# Dataset y DataLoader
# -----------------------
# Repetir H_sim para cada ejemplo
H_sim_batch = H_sim.repeat(X_muones.shape[0], 1)

dataset = TensorDataset(X_muones, H_sim_batch, y_muones)

# Split train/validation (80/20)
n_train = int(0.8 * len(dataset))
n_val = len(dataset) - n_train
train_dataset, val_dataset = random_split(dataset, [n_train, n_val])

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# -----------------------
# Modelo, loss y optimizer
# -----------------------
model = MuonFrequencyNet(n_feat_muon=pulse_samples, n_feat_frec=n_feat_frec).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=lr)

# -----------------------
# Entrenamiento
# -----------------------
train_losses = []
val_losses = []

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for x_muon_batch, x_freq_batch, y_batch in train_loader:
        x_muon_batch = x_muon_batch.to(device)
        x_freq_batch = x_freq_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()
        output = model(x_muon_batch, x_freq_batch)
        loss = criterion(output, y_batch)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * x_muon_batch.size(0)

    epoch_loss = running_loss / n_train
    train_losses.append(epoch_loss)

    # Validación
    model.eval()
    val_running_loss = 0.0
    with torch.no_grad():
        for x_muon_batch, x_freq_batch, y_batch in val_loader:
            x_muon_batch = x_muon_batch.to(device)
            x_freq_batch = x_freq_batch.to(device)
            y_batch = y_batch.to(device)
            val_output = model(x_muon_batch, x_freq_batch)
            val_loss = criterion(val_output, y_batch)
            val_running_loss += val_loss.item() * x_muon_batch.size(0)

    val_epoch_loss = val_running_loss / n_val
    val_losses.append(val_epoch_loss)

    print(f"Epoch {epoch+1}/{epochs} | Train Loss: {epoch_loss:.6f} | Val Loss: {val_epoch_loss:.6f}")

# -----------------------
# Guardar pesos
# -----------------------
os.makedirs("model_weights", exist_ok=True)
torch.save(model.state_dict(), "model_weights/muon_frequency_net.pth")
print("Pesos guardados en 'model_weights/muon_frequency_net.pth'")

# -----------------------
# Plot de pérdida
# -----------------------
plt.figure(figsize=(8,4))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.xlabel("Época")
plt.ylabel("MSE Loss")
plt.title("Entrenamiento de la red")
plt.grid(True)
plt.legend()
plt.show()
```
