# Simulacion de pulsos de muones

El principio del pipeline del proyecto es generar los pulsos de muones con los que voy a entrenar la red neuronal, pero tambien los que voy a usar en mis simulaciones lentas de LTSpice para obtener sus respectivas señales binarias de output.

Hablando de las funciones que estan aca, estas provienen de mi trabajo en Labo6/7 y responden a como son las señales de muones (una concatenacion con offset de pulsos de fotones individuales o SPEs) y los fenomenos probablisiticos que suceden en la barra centelladora.

```python
# Define the function
def func_distance(x, A, lambda1, lambda2):
    """Computes the value of the function f(t) that represents a photon with given parameters, only for t >= 0."""
    term1 = A * np.exp(-x/lambda1)
    term2 = (1 - A) * np.exp(-x/lambda2)
    return term1 + term2

def func_signal(t, A1=0.3, A2=23.8, A3=1.59, tau1=3.9, tau2=1.13, tau3=22.20, tau4=0.251):
    """Computes the value of the function f(t) that represents a photon with given parameters, only for t >= 0."""
    t = np.maximum(t, 0)  # Ensures that there are no negative time offsets
    term1 = A1 * (1 - np.exp(-t / tau1))
    term2 = A2 * np.exp(-t / tau2)
    term3 = A3 * np.exp(-t / tau3)
    term4 = np.exp(-t / tau4)
    return term1 * (term2 + term3 + term4)

def generate_offsets(n, scale):
    """Generate n random offsets using an exponential distribution with a given rate."""
    return np.random.exponential(scale=scale, size=n)

def generate_custom_offsets(n, tau1=3.5, tau2=3.7):
    """Generate n random offsets using the custom distribution combining two exponential distributions."""
    scale1 = tau1
    scale2 = tau2
    
    offsets1 = generate_offsets(n, scale1)
    offsets2 = generate_offsets(n, scale2)
    
    combined_offsets = (offsets1 + offsets2)
    return combined_offsets
```

```python
# Primero quiero mostrar como se ve un pulso de muones

def generate_single_muon_pulse(single_distance):
    """Generate and visualize a single muon pulse at a specific distance."""
    
    # Sample parameters for the Poisson process using func_distance
    A = norm.rvs(loc=42.8, scale=1.8)
    lambda1 = norm.rvs(loc=3.78, scale=0.25)
    lambda2 = norm.rvs(loc=0.035, scale=0.001)
    
    # Calculate expected number of photons at this single distance
    expected_photons = func_distance(single_distance, A, lambda1, lambda2)
    
    # Get Poisson sample for this single distance
    n_photons = poisson.rvs(mu=np.abs(expected_photons))
    
    print(f"Distance: {single_distance:.2f} m")
    print(f"Expected photons (μ): {expected_photons:.2f}")
    print(f"Actual photons generated: {n_photons}")
    
    # Generate the signal pulse
    plt.figure(figsize=(12, 8))
    
    # Sample parameters for func_signal
    A1 = norm.rvs(loc=0.3, scale=0.01)
    A2 = norm.rvs(loc=23.8, scale=0.2)
    A3 = norm.rvs(loc=1.59, scale=0.01)
    tau1 = norm.rvs(loc=3.9, scale=0.2)
    tau2 = norm.rvs(loc=1.13, scale=0.02)
    tau3 = norm.rvs(loc=22.20, scale=0.4)
    tau4 = norm.rvs(loc=0.251, scale=0.005)
    
    # Generate time values for the pulse
    t_values = np.linspace(0, 150, 1000)  # More points for better resolution
    summed_signal = np.zeros_like(t_values)
    
    # Generate offsets for all photons at this distance
    if n_photons > 0:
        photon_offsets = generate_custom_offsets(n_photons)
        print(f"Photon time offsets: {photon_offsets}")
        
        # Plot individual photon signals
        for i, offset in enumerate(photon_offsets):
            individual_signal = func_signal(t_values - offset, A1, A2, A3, tau1, tau2, tau3, tau4)/1000
            plt.plot(t_values, individual_signal, 'gray', alpha=0.3, linewidth=1, 
                    label=f'Photon {i+1}' if i < 5 else "")
            summed_signal += individual_signal
    else:
        print("No photons generated for this muon!")
        photon_offsets = np.array([])
    
    # Plot the summed signal
    plt.plot(t_values, summed_signal, 'r-', linewidth=3, label='Summed signal')
    plt.title(f'Single Muon Pulse at {single_distance:.2f} m - {n_photons} Photons')
    plt.xlabel('Time (ns)')
    plt.ylabel('Signal Amplitude')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.ylim(bottom=0)
    
    # Add text box with information
    textstr = '\n'.join((
        f'Distance: {single_distance:.2f} m',
        f'Expected μ: {expected_photons:.2f}',
        f'Generated: {n_photons} photons',
        f'Max amplitude: {np.max(summed_signal):.4f}'
    ))
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
    plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=10,
             verticalalignment='top', bbox=props)
    
    plt.tight_layout()
    plt.show()
    
    # Print pulse characteristics
    if n_photons > 0:
        max_amplitude = np.max(summed_signal)
        pulse_area = np.trapezoid(summed_signal, t_values)
        pulse_duration = t_values[np.where(summed_signal > max_amplitude * 0.01)[0][-1]] - t_values[np.where(summed_signal > max_amplitude * 0.01)[0][0]]
    
    return t_values, summed_signal, n_photons, photon_offsets

# Example usage for a single muon at a specific distance
single_distance = 2.5  # Choose a specific distance in meters

print("Generating single muon pulse at a specific distance...")
t_values, pulse_signal, n_photons, offsets = generate_single_muon_pulse(single_distance)
```

```python
# Con esta funcion me formo mi señal entera de muones concatenandolos

def concatenate_and_export_photon_signals(x_values, times, time_gap_ns=500):
    all_t_values = []
    all_summed_values = []
    current_time = 0

    # Sample parameters for the Poisson process using func_distance
    A = norm.rvs(loc=42.8, scale=1.8)
    lambda1 = norm.rvs(loc=3.78, scale=0.25)
    lambda2 = norm.rvs(loc=0.035, scale=0.001)
    
    y_values = func_distance(x_values, A, lambda1, lambda2)

    # Generate Poisson samples for each x_value, times times
    poisson_samples = poisson.rvs(mu=np.abs(y_values)[:, np.newaxis], size=(len(x_values), times))

    print("Poisson sampled points:", poisson_samples)

    # Plot the curve and Poisson samples
    plt.figure(figsize=(14, 8))
    plt.plot(x_values, y_values, 'o-', label=f'func_distance with A={A:.2f}, λ1={lambda1:.2f}, λ2={lambda2:.2f}', color='black')
    
    # Plotting each set of samples
    for i in range(times):
        plt.scatter(x_values, poisson_samples[:, i], label=f'Poisson sample set {i+1}', zorder=5)

    plt.title('Function Distance Plot with Poisson Samples')
    plt.xlabel('Distance (m)')
    plt.ylabel('PEs')
    plt.grid(True)
    plt.legend()
    plt.show()

    for index, n_samples in enumerate(poisson_samples):
        for n in n_samples:
            
            #print(n)
            
            # Generate custom offsets for the signal
            random_offsets = generate_custom_offsets(n)
        
            # Generate time values and initialize summed values
            t_values = np.linspace(0, 150, 1000)
            summed_values = np.zeros_like(t_values)

            # Sample parameters for func_signal
            A1 = norm.rvs(loc=0.3, scale=0.01)
            A2 = norm.rvs(loc=23.8, scale=0.2)
            A3 = norm.rvs(loc=1.59, scale=0.01)
            tau1 = norm.rvs(loc=3.9, scale=0.2)
            tau2 = norm.rvs(loc=1.13, scale=0.02)
            tau3 = norm.rvs(loc=22.20, scale=0.4)
            tau4 = norm.rvs(loc=0.251, scale=0.005)


            for offset in random_offsets:
                individual_values = func_signal(t_values - offset, A1, A2, A3, tau1, tau2, tau3, tau4)/1000
                summed_values += individual_values

            t_values = t_values * (10 ** -9)

            t_values += current_time
            all_t_values.extend(t_values)
            all_summed_values.extend(summed_values)
            current_time = t_values[-1] + time_gap_ns * 1e-9  # Convert ns to s for the time gap

    # Save the concatenated data to a file
    filename = f"combined_photon_data_{times}_times.txt"
    with open(filename, 'w') as file:
        for t, value in zip(all_t_values, all_summed_values):
            file.write(f"{t}, {value}\n")

    return all_t_values, all_summed_values, poisson_samples
```

```python
# Iniciar el cronómetro
start_time = time.time()

# Example usage with x_values
x_values = np.linspace(0.4, 4.9, 500)  # Example x_values array
time_gap_ns = 10000  # Example time gap
times = 20

all_t_values, all_summed_values, poisson_samples = concatenate_and_export_photon_signals(x_values, times, time_gap_ns)

# Tiempo transcurrido hasta este punto
mid_time = time.time()
print(f"\nTiempo de simulacion: {mid_time - start_time:.4f} segundos")
```

# Señales de salida y binarias - Matcheo de datos


Con los pulsos que genere ahora los paso por la simulacion de la electronica del LTSpice donde tengo señales cuadradas que tengo que procesar para obtener las correspondencias binarias de cada una



Asi es como se ve el programa de simulacion, la electronica y el output; aclarando que lo azul son  las señales cuadradas de interes:



```python
@njit
def compute_pulse_widths(time_array, vout_array, threshold, time_divisor):
    pulse_widths = []
    pulse_start_time = None

    for i in range(1, len(vout_array)):
        if vout_array[i] > threshold and (pulse_start_time is None or vout_array[i-1] <= threshold):
            # Rising edge detected
            pulse_start_time = time_array[i]
        elif vout_array[i] <= threshold and pulse_start_time is not None and vout_array[i-1] > threshold:
            # Falling edge detected
            pulse_end_time = time_array[i]
            # Calculate pulse width
            width = (pulse_end_time - pulse_start_time) / time_divisor
            pulse_widths.append(np.floor(width))  # Floor to nearest lower integer
            pulse_start_time = None

    return pulse_widths
```

```python
# -----------------------
# Configuración
# -----------------------
pulse_samples = 1000        # cantidad de muestras por pulso (remuestreo para que los inputs de la NN sean iguales)
threshold = 0.2             # umbral para detectar pulsos cuadrados
time_divisor = 3.125e-9     # divisor de tiempo de la señal cuadrada
num_muones = 10000          # número total esperado de muones

# -----------------------
# Cargar archivos
# -----------------------
analog_file = r"C:\Users\smili\OneDrive\Documents\GitHub\Sistemas Dinamicos - Archivos Grandes\10k_highres_muon_NN_train.txt"
square_file = r"C:\Users\smili\OneDrive\Documents\GitHub\Sistemas Dinamicos - Archivos Grandes\10k_highres_BINARY_LTSpice_NN_train.txt"

square_data = pd.read_csv(square_file, sep="\t", header=0)
square_time = square_data['time'].values.astype(np.float64)
square_values = square_data['V(out)'].values.astype(np.float64)

analog_data = pd.read_csv(analog_file, sep=",", header=None, names=["time","value"])
analog_time = analog_data['time'].values.astype(np.float64)
analog_values = analog_data['value'].values.astype(np.float64)

# -----------------------
# Detectar pulsos cuadrados
# -----------------------
pulse_widths_square = compute_pulse_widths(square_time, square_values, threshold, time_divisor)

# Tiempos de inicio de cada pulso cuadrado
pulse_starts = [square_time[0]]
for i in range(1, len(square_values)):
    if square_values[i] > threshold and square_values[i-1] <= threshold:
        pulse_starts.append(square_time[i])

# -----------------------
# Segmentar pulsos analógicos y asignar binarios
# -----------------------
X = []
y = []
last_square_idx = 0

approx_points_per_pulse = len(analog_values) // num_muones

for i in range(num_muones):
    start_idx = i * approx_points_per_pulse
    end_idx = start_idx + approx_points_per_pulse
    pulse_segment = analog_values[start_idx:end_idx]

    # Interpolar a tamaño fijo
    pulse_segment = np.interp(
        np.linspace(0, len(pulse_segment)-1, pulse_samples),
        np.arange(len(pulse_segment)),
        pulse_segment
    )
    X.append(pulse_segment)

    # Asignar binario seguro
    if last_square_idx < len(pulse_widths_square):
        y.append(pulse_widths_square[last_square_idx])
        last_square_idx += 1
    else:
        y.append(0)

X = np.array(X)
y = np.array(y)
print(f"Dataset final: {X.shape[0]} pulsos, cada uno con {X.shape[1]} muestras.")

# -----------------------
# Chequear cantidad de ceros
# -----------------------
num_zeros = np.sum(y == 0)
num_detected = len(y) - num_zeros
print(f"Pulsos totales: {len(y)}")
print(f"Pulsos con binario 0: {num_zeros}")
print(f"Pulsos con binario >0: {num_detected}")
print(f"Porcentaje de ceros: {100*num_zeros/len(y):.2f}%")

# -----------------------
# Guardar dataset
# -----------------------
np.savez("muon_dataset_segmented_v3.npz", X=X, y=y)
print("Dataset guardado como 'muon_dataset_segmented_v3.npz'.")

# -----------------------
# Debug: primer pulso + 3 al azar
# -----------------------
indices = [0] + random.sample(range(1, len(X)), 3)

for i in indices:
    plt.figure(figsize=(12,4))
    
    # Pulso analógico segmentado
    plt.plot(X[i]*30, label="Pulso analógico segmentado")
    
    # Título con binario asignado
    plt.title(f"Pulso {i} | Binario = {y[i]}")
    plt.xlabel("Muestra")
    plt.ylabel("Amplitud")

    # Graficar señal cuadrada en la ventana aproximada
    pulse_start_time = analog_time[i * approx_points_per_pulse]
    pulse_end_time = analog_time[min((i+1) * approx_points_per_pulse, len(analog_time)-1)]
    mask = (square_time >= pulse_start_time) & (square_time <= pulse_end_time)
    
    square_window_time = square_time[mask]
    square_window_values = square_values[mask]
    
    if len(square_window_time) > 0:
        square_window_interp = np.interp(
            np.linspace(0, len(square_window_time)-1, X[i].shape[0]),
            np.arange(len(square_window_time)),
            square_window_values
        )
        plt.plot(square_window_interp, label="Señal cuadrada (interpolada)")

    plt.legend()
    plt.grid(True)
    plt.show()
```

# Preparar espectros en frecuencia

```python
# -----------------------
# CONFIG
# -----------------------
vna_s2p_path = r'C:\Users\smili\OneDrive\Documents\GitHub\Laboratorio_6_7_ITeDA\Mediciones\Mediciones CITI VNA\2024-12-03\Citi1-P12-FS\ganancia_0.s2p'
pitaya_csv_path = r'C:\Users\smili\OneDrive\Documents\GitHub\Laboratorio_6_7_ITeDA\Mediciones\Mediciones Pitaya\29.11.2024\full_chain_gain0.csv'
simulated_txt_path = r'C:\Users\smili\OneDrive\Documents\GitHub\Laboratorio_6_7_ITeDA\Etapa Final de Proyecto\bode_circuito_completo_con_capacitores_en_preamp_inversor_v2.txt'

output_dir = 'Espectros'
os.makedirs(output_dir, exist_ok=True)

match_freq = 300e3       # frecuencia de pegado Pitaya -> VNA
f_min_mask = 1e3         # Hz, por debajo -> mask_value
f_max_mask = 5e8         # Hz, por encima -> mask_value
mask_value = -100.0      # dB
n_points = 5000          # puntos log-space para ambos espectros

# -----------------------
# FUNCIONES
# -----------------------
def read_s2p_file(path):
    with open(path, 'r') as f:
        lines = [l for l in f if not (l.startswith('#') or l.startswith('!'))]
    data = [l.strip().split() for l in lines]
    df = pd.DataFrame(data).apply(pd.to_numeric)
    freq = df.iloc[:, 0].to_numpy()
    s21  = df.iloc[:, 3].to_numpy()
    return freq, s21

def read_pitaya_csv(path):
    df = pd.read_csv(path)
    df.columns = df.columns.str.strip()
    freq = df['Frequency [Hz]'].to_numpy(float)
    amp  = df['Amplitude [dB]'].to_numpy(float)
    return freq, amp

# -----------------------
# LECTURA
# -----------------------
print("Leyendo archivos...")
vna_freq, vna_s21 = read_s2p_file(vna_s2p_path)
pitaya_freq, pitaya_amp = read_pitaya_csv(pitaya_csv_path)

df_sim = pd.read_csv(
    simulated_txt_path, sep="\t", skiprows=1, names=['Frequency','V(out)'], encoding='ISO-8859-1'
)
df_sim['Attenuation'] = df_sim['V(out)'].apply(lambda x: float(x.split(',')[0][1:].replace('dB','').strip()))
sim_freq, sim_att = df_sim['Frequency'].to_numpy(), df_sim['Attenuation'].to_numpy()

# -----------------------
# STITCHING PITAYA + VNA
# -----------------------
vna_val_at_match = np.interp(match_freq, vna_freq, vna_s21)
pitaya_val_at_match = np.interp(match_freq, pitaya_freq, pitaya_amp)
pitaya_offset = vna_val_at_match - pitaya_val_at_match
print(f"Offset calculado para Pitaya en {match_freq} Hz = {pitaya_offset:.3f} dB (VNA @ match = {vna_val_at_match:.3f}, Pitaya antes = {pitaya_val_at_match:.3f})")


pitaya_amp_matched = pitaya_amp + pitaya_offset

mask_pitaya = pitaya_freq <= match_freq
mask_vna    = vna_freq > match_freq
real_freq_concat = np.concatenate([pitaya_freq[mask_pitaya], vna_freq[mask_vna]])
real_att_concat  = np.concatenate([pitaya_amp_matched[mask_pitaya], vna_s21[mask_vna]])

# Ordenar por frecuencia
idx_sort = np.argsort(real_freq_concat)
real_freq_concat = real_freq_concat[idx_sort]
real_att_concat  = real_att_concat[idx_sort]

# -----------------------
# REMUESTREAR EN LOG-SPACE
# -----------------------
freq_grid = np.logspace(np.log10(f_min_mask), np.log10(f_max_mask), n_points)
sim_interp  = np.interp(freq_grid, sim_freq, sim_att, left=mask_value, right=mask_value)
real_interp = np.interp(freq_grid, real_freq_concat, real_att_concat, left=mask_value, right=mask_value)

# -----------------------
# MASKING EXPLÍCITO
# -----------------------
sim_interp[freq_grid < f_min_mask] = mask_value
sim_interp[freq_grid > f_max_mask] = mask_value
real_interp[freq_grid < f_min_mask] = mask_value
real_interp[freq_grid > f_max_mask] = mask_value

# -----------------------
# NORMALIZAR PICO A 0 dB
# -----------------------
sim_interp  -= np.max(sim_interp)
real_interp -= np.max(real_interp)

# -----------------------
# GUARDAR SOLO .npz
# -----------------------
np.savez(os.path.join(output_dir,'spectrum_simulated.npz'), freq=freq_grid, attenuation_db=sim_interp)
np.savez(os.path.join(output_dir,'spectrum_real.npz'), freq=freq_grid, attenuation_db=real_interp)
print("Espectros guardados en formato .npz")

# -----------------------
# OPCIONAL: visualización rápida
# -----------------------
plt.figure(figsize=(9,5))
plt.semilogx(freq_grid, sim_interp, label='Simulado', linewidth=1)
plt.semilogx(freq_grid, real_interp, label='Real (Pitaya+VNA stitched)', linewidth=1)
plt.axvline(match_freq, color='k', linestyle='--', label=f'Pegado @ {match_freq/1e3:.0f} kHz')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Atenuación Normalizada (dB, pico=0)')
plt.title('Simulado vs Real (stiched, masked, normalizado)')
plt.grid(True, which='both', ls=':')
plt.legend()
plt.tight_layout()
plt.show()
```
